# Plotting Sigmoid Function - 2D

```python
def sigmoid(x, w, b):
    return 1/(1 + np.exp(-(w*x + b)))

def plot_sgimoid_2d(w, b):
    X = np.linspace(-10,10,100)
    Y = sigmoid(X,w,b)
    plt.plot(X,Y)
    plt.show()

interact(plot_sgimoid_2d,w=(-2,2,0.1),b=(-2,2,0.1));
```

# Plotting Sigmoid Function - 3D

```python
def sigmoid_2d(x1, x2, w1, w2, b):
    return 1/(1 + np.exp(-(w1*x1 + w2*x2 + b)))

def plot_sgimoid_3d(w1, w2, b):
    X1 = np.linspace(-10,10,100)
    X2 = np.linspace(-10,10,100)
    X1, X2 = np.meshgrid(X1, X2)
    Y = sigmoid_2d(X1,X2,w1,w2,b)
    
    fig = plt.figure(figsize=(7.5,7.5))
    ax = plt.axes(projection='3d')
    ax.plot_surface(X1, X2, Y, cmap='viridis')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('Y');
    
    ax.view_init(40,35);

interact(plot_sgimoid_3d,w1=(-2,2,0.1),w2=(-2,2,0.1),b=(-2,2,0.1));
```

# Class for Sigmoid Neuron

```python
class SigmoidNeuron:
    
    def __init__(self):
        self.w = None
        self.b = None
    
    def perceptron(self,x):
        return np.dot(x,self.w.T) + self.b
    
    def sigmoid(self,x):
        return 1.0/(1.0 + np.exp(-x))
    
    def grad_w(self,x,y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y)*y_pred*(1 - y_pred)*x
    
    def grad_b(self,x,y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y)*y_pred*(1 - y_pred)
    
    def fit(self,X,Y,epochs=1,learning_rate=1,intialise=True):
        # intialize w,b
        if intialise:
            self.w = np.random.randn(1,X.shape[1])
            self.b = 0
        
        for i in range(epochs):
            dw = 0
            db = 0
            for x,y in zip(X,Y):
                dw += self.grad_w(x,y)
                db += self.grad_b(x,y)
            self.w -= learning_rate * dw
            self.b -= learning_rate * db
```

# Load Real World Dataset

```python
data = pd.read_csv('mobile_cleaned.csv')
data.head()
```

```python
X = data.drop('Rating',axis=1).values
Y = data['Rating'].values
threshold = 4.2
data['Class'] = (data['Rating'] >= threshold).astype('int')
print(data['Class'].value_counts(normalize=True))
Y_binarised = data['Class'].values
```

```python
scaled_threshold = min_max_scaler.transform(np.asarray([threshold]).reshape(-1,1))[0][0]
scaled_threshold
```

```python
Y_binarised_train = (Y_scaled_train > scaled_threshold).astype('int').ravel()
Y_binarised_test = (Y_scaled_test > scaled_threshold).astype('int').ravel()
```

# Class of Sigmoid Neuron and Train on real dataset

```python
class SigmoidNeuron:
    
    def __init__(self):
        self.w = None
        self.b = None
    
    def perceptron(self,x):
        return np.dot(x,self.w.T) + self.b
    
    def sigmoid(self,x):
        return 1.0/(1.0 + np.exp(-x))
    
    def grad_w(self,x,y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y)*y_pred*(1 - y_pred)*x
    
    def grad_b(self,x,y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y)*y_pred*(1 - y_pred)
    
    def fit(self,X,Y,epochs=1,learning_rate=1,intialise=True,display_loss=False):
        # intialize w,b
        if intialise:
            self.w = np.random.randn(1,X.shape[1])
            self.b = 0
            
        if(display_loss):
            loss = {}
        
        for i in tqdm_notebook(range(epochs),total=epochs,unit='epoch'):
            dw = 0
            db = 0
            for x,y in zip(X,Y):
                dw += self.grad_w(x,y)
                db += self.grad_b(x,y)
            self.w -= learning_rate * dw
            self.b -= learning_rate * db
            
            if(display_loss):
                Y_pred = self.sigmoid(self.perceptron(X))
                loss[i] = mean_squared_error(Y_pred,Y)
            
        if(display_loss):
            plt.plot(loss.values())
            plt.xlabel('Epochs')
            plt.ylabel('Loss')

    def predict(self,X):
        Y_pred = []
        for x in X:
            y_pred = self.sigmoid(self.perceptron(x))
            Y_pred.append(y_pred)
        return np.asarray(Y_pred)
```

```python
sn = SigmoidNeuron()
sn.fit(X_scaled_train,Y_scaled_train,epochs=2000,learning_rate=0.015,display_loss=True)
Y_pred_train = sn.predict(X_scaled_train)
Y_pred_test = sn.predict(X_scaled_test)

Y_pred_binarised_train = (Y_pred_train > scaled_threshold).astype('int').ravel()
Y_pred_binarised_test = (Y_pred_test > scaled_threshold).astype('int').ravel()

accuracy_train = accuracy_score(Y_binarised_train,Y_pred_binarised_train)
accuracy_test = accuracy_score(Y_binarised_test,Y_pred_binarised_test)
print(accuracy_train,accuracy_test)
```

# For one feature - Sigmoid Neuron - Mean Squared Error

```python
X = [0.5,2.5]
Y = [0.2,0.9]

def f(w, b, x):
    return 1/(1 + np.exp(-(w*x + b)))

def error(w,b):
    err = 0.0
    for x,y in zip(X,Y):
        fx = f(w, b, x)
        err += 0.5*(fx -y)**2
    return err

def grad_b(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y)*fx*(1-fx)

def grad_w(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y)*fx*(1-fx)*x

def do_gradient_descent():
    w, b, eta = -2,-2,1.0
    max_epochs = 100
    for i in range(max_epochs):
        dw,db = 0,0
        for x,y in zip(X,Y):
            dw += grad_w(w,b,x,y)
            db += grad_b(w,b,x,y)
        w = w - eta*dw
        b = b - eta*db
```

# For one feature - Sigmoid Neuron - With Cross Entropy Loss Function

```python
import math

X = [0.5,2.5]
Y = [0.2,0.9]

def f(w, b, x):
    return 1/(1 + np.exp(-(w*x + b)))

def error(w,b):
    err = 0.0
    for x,y in zip(X,Y):
        fx = f(w, b, x)
        err += -[(1-y) * math.log(1 - fx,2) + y * math.log(fx,2)]
    return err

def grad_b(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y)

def grad_w(w, b, x, y):
    fx = f(w, b, x)
    return (fx - y)*x

def do_gradient_descent():
    w, b, eta = -2,-2,1.0
    max_epochs = 100
    for i in range(max_epochs):
        dw,db = 0,0
        for x,y in zip(X,Y):
            dw += grad_w(w,b,x,y)
            db += grad_b(w,b,x,y)
        w = w - eta*dw
        b = b - eta*db
```

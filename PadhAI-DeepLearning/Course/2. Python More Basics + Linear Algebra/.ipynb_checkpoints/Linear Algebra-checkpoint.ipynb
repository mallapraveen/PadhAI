```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting Vectors

```python
plt.quiver(0,0,3,4)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='red')
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1,color='green')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
def plot_vectors(vecs):
    colors = ['r','g','b','y']
    i = 0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)])
        i+=1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,3,-4),(0,0,-3,-4)])
```

```python
vecs = [(0,0,3,4),(0,0,-3,4),(0,0,3,-4),(0,0,-3,-4)]
```

```python
plot_vectors([vecs[0],vecs[3]])
```

```python
vecs[0] + vecs[2]
```

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,3,-4]),np.asarray([0,0,-3,-4])]
```

```python
plot_vectors([vecs[0],vecs[3]])
```

### Vector Addition and Subtraction

```python
plot_vectors([vecs[0]+vecs[2]])
```

```python
plot_vectors([vecs[0]+vecs[2],vecs[1]+vecs[3],vecs[1]])
```

```python
plot_vectors([vecs[0]+vecs[0],vecs[0]])
```

```python
plot_vectors([vecs[0]-vecs[1]])
```

### Vector Multiplication

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-4]),np.asarray([0,0,4,-1])]
```

```python
plot_vectors(vecs)
```

```python
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

```python
np.dot(a,b)
```

### Project of a on b

```python
vec_a_b = np.dot(a,b)/(np.linalg.norm(b)*np.linalg.norm(b)) * b
print(vec_a_b)
```

```python
plot_vectors([np.asarray([0,0,3,4]),np.asarray([0,0,4,-1]),np.asarray([0,0,1.88235294,-0.47058824])])
```

```python
vec_b_a = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(a)) * a
print(vec_b_a)
```

# Central Limit Theorem

## Normal Distribution

```python
k = 10000
n = 1000

mu, sigma = 0, 1 # mean and standard deviation
x = np.random.normal(mu, sigma, k)

s = np.random.choice(x,n)
b = []
for i in range(k):
    s = np.random.choice(x,n)
    b.append(s)

b = np.array(b)
b = b.sum(axis=1)
```

```python
sns.kdeplot(x)
```

```python
plt.hist(b,bins=100)
plt.show()
```

## Weibull Distribution

```python
k = 10000
n = 1000

mu, sigma = 0, 1 # mean and standard deviation
x = np.random.weibull(k,n)

s = np.random.choice(x,n)
b = []
for i in range(k):
    s = np.random.choice(x,n)
    b.append(s)

b = np.array(b)
b = b.sum(axis=1)
```

```python
sns.kdeplot(x)
```

```python
plt.hist(b,bins=100)
plt.show()
```

# Plot Variance with respect to size of sample

## Exerise 1

```python
np.arange(1,50)
```

```python
k = 10000
n = np.arange(1,50)
std = []
for i in n:
    std.append(np.std(np.mean(np.random.randint(1,7,(k,i)),axis=1)))
```

```python
sns.lineplot(n,std)
plt.show()
```

# Plot Variance with respect to mu(Mean) and sigma(variance)

## Exercise - 2

```python
n = 10
k = 1000
mu = np.arange(-1,1.1,0.1)
mu_var = []
sigma = np.arange(0,11)
sigma_var = []
for i in mu:
    a = i + np.random.randn(k,n)
    mu_var.append(np.var(a.mean(axis=1)))
    
for i in sigma:
    a = i*np.random.randn(k,n)
    sigma_var.append(np.var(a.mean(axis=1)))
```

```python
sns.lineplot(mu,mu_var)
plt.ylim(0.05,0.15)
plt.show()
```

```python
sns.lineplot(sigma,sigma_var)
plt.show()
```
